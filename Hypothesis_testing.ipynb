```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use("fivethirtyeight")
import warnings
warnings.filterwarnings('ignore')
from scipy.stats import ttest_ind
```

```python
cereal=pd.read_csv("C:/Users/samata.suresh.naik/Documents/Stats and ML/Stats/Class 8/cereal.csv")
```

```python
cereal.head()
```

```python
hot_cereal=cereal.loc[cereal['type'] == 'H',:] # defining hot cereal data frame
cold_cereal= cereal.loc[cereal['type'] == 'C', :] # defining cold cereal dataframe
```

```python
ttest_ind(cold_cereal['sugars'],hot_cereal['sugars'], equal_var=False)
```

```python
ttest_ind(cold_cereal['calories'],hot_cereal['calories'], equal_var=False)
```

```python
#lets visualize
ax= plt.subplots(figsize=(18,8)) # make plot larger
sns.distplot(cold_cereal['sugars'],bins=10,hist=True, label='cold') # plot cold cereal sugar distribution
sns.distplot(hot_cereal['sugars'],bins=10,hist=True, label='cold') # plot hot cereal sugar distribution
plt.legend()
```
