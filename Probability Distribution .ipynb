```python
#poisson
import matplotlib.pyplot as plt

gaussian=np.random.normal(0.5,2*np.sqrt(2*np.log(2)),1000)
poisson=np.random.poisson(0.5,1000)

plt.hist(gaussian)
plt.hist(poisson)
plt.show()
```
